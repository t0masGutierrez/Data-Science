```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Christmas Tree Sales.csv")
df
```

```python
# Scale #s
df["Number of trees sold"] /= 1000000
df["Sales"] /= 1000000000
```

```python
# Summary Statistics
df.describe()
```

```python
# Number of trees sold by Type of tree
real = df["Type of tree"] == "Real tree"
real = df.loc[real, "Number of trees sold"].sum()

fake = df["Type of tree"] == "Fake tree"
fake = df.loc[fake, "Number of trees sold"].sum()

plt.pie([fake, real], labels = ["Fake trees", "Real trees"], autopct = "%.1f%%")
plt.title("Number of trees sold by Type of tree")
plt.show()
```

```python
# Number of trees sold by Year and Type of tree
df.pivot_table(values = "Number of trees sold", index = "Year", columns = "Type of tree").plot.bar()
plt.ylabel("# of trees sold (Millions)")
plt.title("Number of trees sold by Year and Type of tree")
plt.show()
```

```python
# Sales by Year and Type of tree
df.pivot_table(values = "Sales", index = "Year", columns = "Type of tree").plot(kind = "barh")
plt.ylabel("Sales (Billions)")
plt.title("Sales by Year and Type of tree")
plt.show()
```

```python
# Average Tree Price by Year and Type of tree
df.pivot_table(values = "Average Tree Price", index = "Year", columns = "Type of tree").plot.line()
plt.ylabel("Average Tree Price")
plt.title("Average Tree Price by Year and Type of tree")
plt.show()
```
